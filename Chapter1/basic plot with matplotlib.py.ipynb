```python
# Basic Plot with Matplotlib

# Data Visualization
# Data Structures
# Control Structures
# Case Study

# Data Visualization
# ------------------------

import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year,pop,color="green")
plt.show()
```

```python
# Scatter plot

import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year,pop,color="red")
plt.show()
```

```python
# Histogram

import matplotlib.pyplot as plt

# help(plt.hist)

values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(values,bins=3)
plt.show()
```

```python
# Basic plot

import matplotlib.pyplot as plt
year = [1950, 1951, 1952, 2100]
pop = [2.538, 2.57, 2.62,10.85]

plt.plot(year,pop)
plt.show()
```

```python
# Xlabel & ylabel
import matplotlib.pyplot as plt
year = [1950, 1951, 1952, 2100]
pop = [2.538, 2.57, 2.62,10.85]

plt.plot(year,pop,color="green")
plt.xlabel("YEAR", color="red")
plt.ylabel("POPULATION", color="red")
plt.show()
```

```python
# Title
import matplotlib.pyplot as plt
year = [1950, 1951, 1952, 2100]
pop = [2.538, 2.57, 2.62,10.85]

plt.plot(year,pop,color="green")
plt.xlabel("YEAR", color="red")
plt.ylabel("POPULATION", color="red")
plt.title("World Population", color="green")
plt.show()
```

```python
# Ytikcs
import matplotlib.pyplot as plt
year = [1950, 1951, 1952, 2100]
pop = [2.538, 2.57, 2.62,10.85]

plt.plot(year,pop,color="green")
plt.xlabel("YEAR", color="red")
plt.ylabel("POPULATION", color="red")
plt.title("World Population", color="green")
plt.yticks([0,2,4,6,8,10])
plt.show()
```

```python
# Add historical data

import matplotlib.pyplot as plt 
year = [1950, 1951, 1952, 2100] 
pop = [2.538, 2.57, 2.62, 10.85]

# Add more data
year = [1800,1850,1900] + year
pop = [1.0, 1.262, 1.650] + pop

plt.plot(year,pop)
plt.title("WORLD POPULATION DATA")
plt.xlabel("YEAR")
plt.ylabel("POPULATION")
plt.show()
```
